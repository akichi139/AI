```python
from gtts import gTTS
from playsound import playsound
import os
from IPython.display import Audio, display
```

```python
def create_audio_dataset(texts, output_folder="audio_dataset", language='th', slow=False):
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    for i, text in enumerate(texts, 1):
        tts = gTTS(text=text, lang=language, slow=slow)
        audio_filename = f"audio_{i}.wav"
        audio_path = os.path.join(output_folder, audio_filename)
        tts.save(audio_path)
        print(f"Text {i}/{len(texts)} processed. Saved to {audio_path}")

# Example usage:
texts = [
    "สวัสดีค่ะผู้สร้าง",
    "เช้านี้อากาศสดใสดีนะคะ",
    "ผู้สร้างหิวข้าวมั้ยคะหรือหิวตีนมากกว่าคะ",
    "ลองตีนดิฉันมั้ยคะ",
    "รับรองว่าอร่อยแน่นอนค่ะ"
]

create_audio_dataset(texts)
```

```python
def play_audio_dataset(audio_folder="audio_dataset"):
    for file in os.listdir(audio_folder):
        if file.endswith(".wav"):
            audio_path = os.path.join(audio_folder, file)
            print(f"Playing: {audio_path}")
            display(Audio("/content/"+audio_path, autoplay=True))
            # playsound("/content/"+audio_path)
```
