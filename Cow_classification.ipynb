```python
import pandas as pd
import numpy as np

df = pd.read_csv('/content/cow.csv')
```

## Clean data

### Drop hospital number column

```python
df = df.drop(columns=['hospital_number'], axis=1)
df
```

### Finding numeric columns

```python
num_cols = df.select_dtypes(include=["float", "int"]).columns
num_df = df[num_cols]

num_df
```

### Create X set

```python
x = df.drop(["outcome Class"], axis=1)
x
```

### Create Y set

```python
y = df["outcome Class"]
y
```

### Fillna with 0 in numeric columns

```python
x[num_cols]=x[num_cols].fillna(0)
x[num_cols]
```

### Fillna with missing in nominal columns

```python
x=x.fillna("missing")
x
```

## Pre-processing data

### Standard Scaling

```python
from sklearn.preprocessing import StandardScaler

col_scaler_mapping = {}
for col in num_cols:
    scaler = StandardScaler()
    x[[col]] = scaler.fit_transform(x[[col]])
    col_scaler_mapping[col] = scaler
```

### Encoding labels

```python
from sklearn import preprocessing

le = preprocessing.LabelEncoder()
for col in x.columns:
  if(col not in num_cols):
    x[col]=le.fit_transform(x[col])
    print(le.classes_)

y=le.fit_transform(y)
print(le.classes_)
```

```python
x
```

```python
y
```

## Split data

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
```

## Feature selection


### Chi Square

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2

X_train_nominals = X_train.loc[:, ~X_train.columns.isin(num_cols)]
chi2_selector = SelectKBest(chi2, k=8)
chi2_selector.fit_transform(X_train_nominals, y_train)
chi2_top_features = X_train_nominals.columns[chi2_selector.get_support()]
scores = chi2_selector.scores_

chi2_top_features
```

### LDA (Linear Discriminant Analysis)

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.datasets import load_iris

# Initialize and fit the LDA model
lda = LinearDiscriminantAnalysis(n_components=1)
lda.fit(X_train[num_cols], y_train)

# Transform the data using the LDA model
X_train_lda = lda.transform(X_train[num_cols])
X_test_lda = lda.transform(X_test[num_cols])

print("Train column: ", len(X_train_lda[0]))
print("Test column: ", len(X_test_lda[0]))
```

### RFECV (Recursive feature elimination with cross-validation) using Random Forest as estimator

```python
from sklearn import ensemble
from sklearn.feature_selection import RFECV
from sklearn.metrics import classification_report

clf = ensemble.RandomForestClassifier()

rfe = RFECV(estimator=clf)
rfe.fit(X_train, y_train)

rfecv_top_features = X_train.columns[rfe.support_]

print(rfecv_top_features)
```

## Models

### Logistic Regression using Chi square as feature selection method

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report

log_reg = LogisticRegression()
model = log_reg.fit(X_train[chi2_top_features], y_train)
y_true = y_test
y_pred = model.predict(X_test[chi2_top_features])

print(classification_report(y_true, y_pred, zero_division=0))
```

### Logistic Regression using LDA  as feature selection method

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report

log_reg = LogisticRegression()
model = log_reg.fit(X_train_lda, y_train)
y_true = y_test
y_pred = model.predict(X_test_lda)

print(classification_report(y_true, y_pred, zero_division=0))
```

### Decision Tree using RFECV as feature selection method

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report

dtc = DecisionTreeClassifier()
model = dtc.fit(X_train[rfecv_top_features],y_train)
y_true = y_test
y_pred = model.predict(X_test[rfecv_top_features])

print(classification_report(y_true, y_pred, zero_division=0))
```

```python
from sklearn.model_selection import cross_val_score

scores = cross_val_score(model, X_train[rfecv_top_features], y_train, cv=10)
print(scores)
print("Mean: ", np.mean(scores))
```
